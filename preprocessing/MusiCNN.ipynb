```python
import numpy as np

import warnings
warnings.filterwarnings('ignore')

pool4 = np.full( (400, 220, 6, 128)  , np.nan)
pool5 = np.full( (400, 40, 128), np.nan)
taggram_array = np.full( (400, 20, 50) , np.nan)

genres = ["classical", "electronic", "pop", "rock"]

i = 0

for genre in genres:

    path = "./data/emotifymusic/" + genre

    for file in os.scandir(path):
        if file.path.endswith(".mp3"):
            print("analyzing ", file.name, file.path)
            taggram, tags, features = extractor(f"{path}/{file.name}", model='MTT_vgg', extract_features=True)
            
            taggram_shape = np.shape(taggram)
            features_shape_pool4 = np.shape(features["pool4"])
            features_shape_pool5 = np.shape(features["pool5"])
            
            
            print(taggram_shape)
            print(features_shape_pool4)
            print(features_shape_pool5)
            
            
            if taggram_shape[0] < 20:
                taggram = np.append(taggram, np.full( (20-taggram_shape[0], 50), np.nan), axis=0)
                
            if features_shape_pool4[0] < 220:
                features["pool4"] = np.append(features["pool4"], np.full((220 - features_shape_pool4[0], 6, 128), np.nan),
                                             axis=0)
                
            if features_shape_pool5[0] < 40:
                features["pool5"] = np.append(features["pool5"], np.full((40 - features_shape_pool5[0], 128), np.nan),
                                             axis=0)
                
            taggram_array[i] = taggram[:20, :]
            pool4[i] = features["pool4"][:220, :, :]
            pool5[i] = features["pool5"][:40, :, ]
            
            i += 1
```

```python
np.save("MusicCNNFeatures/pool4", pool4)
np.save("MusicCNNFeatures/pool5", pool5)

np.save("MusicCNNFeatures/taggram_array", taggram_array)
```

```python
%matplotlib inline
import matplotlib.pylab as plt
import matplotlib.gridspec as gridspec
import numpy as np

def depict_features(features, coordinates, title, aspect='auto', xlabel=True, fontsize=13):
    # plot features in coordinates
    ax = plt.subplot(coordinates) 
    plt.imshow(features.T, interpolation=None, aspect=aspect)
    # set title
    ax.title.set_text(title + ' (' + str(features.shape[1]) + ')' )
    ax.title.set_fontsize(fontsize)
    # y-axis
    ax.get_yaxis().set_visible(False)
    # x-axis
    x_label = np.arange(0, features.shape[0], features.shape[0]//5)
    ax.set_xticks(x_label)
    ax.set_xticklabels(x_label, fontsize=fontsize)
    if xlabel:
        ax.set_xlabel('(time)', fontsize=fontsize)
```

```python
# visualize pool1 features!
gs = gridspec.GridSpec(3, 2) # create a figure having 2 rows and 3 cols.

depict_features(features=features['pool1'][:,:,0],
                coordinates=gs[0, 0],
                title='pool1: feature #0',
                xlabel=False)

depict_features(features=features['pool1'][:,:,35],
                coordinates=gs[1, 0],
                title='pool1: feature #35',
                xlabel=False)

depict_features(features=features['pool1'][:,:,111],
                coordinates=gs[2, 0],
                title='pool1: feature #111',
                xlabel=False)

depict_features(features=features['pool1'][:,:,3],
                coordinates=gs[0, 1],
                title='pool1: feature #3',
                xlabel=False)

depict_features(features=features['pool1'][:,:,39],
                coordinates=gs[1, 1],
                title='pool1: feature #39',
                xlabel=False)

depict_features(features=features['pool1'][:,:,101],
                coordinates=gs[2, 1],
                title='pool1: feature #101',
                xlabel=False)

plt.tight_layout()
plt.show()
```
